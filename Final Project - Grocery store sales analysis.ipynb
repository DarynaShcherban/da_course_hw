# Як залежить прибуток від часу? (рік, місяць, день)

```python
year_profit = df.groupby('Year')[['Profit Margin']].sum()
sns.barplot(x='Year',y='Profit Margin', data=year_profit)
plt.title("Profit by year")
plt.show(block=False)
```

Гарна статистика - із кожним роком прибуток збільшується.

```python
date_profit = df.groupby('Date')[['Profit Margin']].sum()
plt.figure(figsize=(12, 6))
plt.plot(date_profit)
plt.xticks(rotation=90)
plt.title("Profit by day (Line graph)")
plt.show(block=False)
```

не зрозуміло трохи що за скачки особливо вкінці, тому глянемо на іншому графіку

# Що купують в найпопулярніший місяць? (Які товари найпопулярніші та найприбутковіші?)

Проте бачимо трохи вище (bar plot profit by month) що в грудні у нас найбільший прибуток, цікаво дослідити які ж продукти купують люди в цей період (візьмемо тільки 12й місяць)

```python
df_dec = df[(df['Month'] == 12)]
unique_category_counts = df_dec['Category'].value_counts().reset_index()

# Перейменуємо стовпці для більш зрозумілого найменування
unique_category_counts.columns = ['Category', 'Count']

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Count', data=unique_category_counts)
plt.xticks(rotation=90)
plt.title("Quantity of sold goods by category")
plt.show()
```

```python
unique_subcategory_counts = df_dec.groupby(['Category', 'Sub_Category']).size().reset_index(name='Count')
plt.figure(figsize=(12, 6))
sns.barplot(x='Sub_Category', y='Count', hue='Category', width=8, data=unique_subcategory_counts)
plt.xticks(rotation=90)
plt.title("Quantity of sold goods by subcategory")
plt.show()
```

трохи обмежимо дані щоб подивитись найпопулярніші позиції.

```python
unique_subcategory_counts = df_dec.groupby(['Category', 'Sub_Category']).size().reset_index(name='Count')
unique_subcategory_counts = unique_subcategory_counts[unique_subcategory_counts['Count'] > 125]
plt.figure(figsize=(8, 6))
sns.barplot(x='Sub_Category', y='Count', hue='Category', data=unique_subcategory_counts)
plt.xticks(rotation=90)
plt.title("Top subcategories by quantity of sold goods")
plt.show()
```

Побачили що люди купують найбільше у період грудня, та цікаво подивитись який прибуток ми маємо і з чого найбільший.

```python
category_profit_sum = df_dec.groupby(['Category'])[['Profit Margin']].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit Margin', data=category_profit_sum)
plt.xticks(rotation=90)
plt.title("Profit Margin from sold goods by category")
plt.show()
```

```python
subcategory_profit_sum = df_dec.groupby(['Category', 'Sub_Category'])[['Profit Margin']].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x='Sub_Category', y='Profit Margin', hue='Category', width=8, data=subcategory_profit_sum)
plt.xticks(rotation=90)
plt.title("Profit Margin from sold goods by subcategory")
plt.show()
```

```python
subcategory_profit_sum = df_dec.groupby(['Category', 'Sub_Category'])[['Profit Margin']].sum()
subcategory_profit_sum = subcategory_profit_sum[subcategory_profit_sum['Profit Margin'] > 250]
plt.figure(figsize=(10, 6))
sns.barplot(x='Sub_Category', y='Profit Margin', hue='Category', data=subcategory_profit_sum)
plt.xticks(rotation=90)
plt.title("Top subcategories by profit of sold goods")
plt.show()
```

Бачимо що продукцію із підкатегорій "Oral Care" та "skin care"  купують достатню кількість, і вони приносять найбільший прибуток, тож варто зосередити свою увагу на цих продуктах. А от "BBQ Supplies" хоч і найпопулярніша підкатегорія (кількість), проте це не приносить нам значного прибутку, тож можливо варто розглянути інших (дешевших) постачальників, оскільки ці товари затребувані у наших магазинах.

```python
from scipy import stats as st
df_year_profit = df.pivot_table(index='Year', columns='Gender', values='Profit Margin', aggfunc='sum')
res = st.shapiro(df_year_profit)
alpha = 0.05
if(res.pvalue <  alpha):
    print(f"Відхилити Н0 про нормальність розподілу на рівні {alpha}.")
else:
    print(f"Н0 про нормальність розподілу не може бути відхилена на рівні {alpha}.")
```
